# Spearman rank correlation

calculcate Spearman correlation for $\alpha, \beta$, and $\zeta$ parameters in Weber noise condition

for assessment of reliability

Make list for the median values of the parameters for the runs

```python
list_sub_a=[]
list_sub_b=[]
list_sub_z=[]

# loop through subjects
for sub in list_sub:
    
    str_sub = str(sub)
    # get data for subject
    curr_sub = data.loc[str_sub]
    
    # get parameter values
    alpha = curr_sub['alpha'].values.tolist()
    beta  = curr_sub['beta'].values.tolist()
    zeta  = curr_sub['zeta'].values.tolist()
    
    # calculcate median over all runs
    md_a  = np.median(alpha)
    md_b  = np.median(beta)
    md_z  = np.median(zeta)
    
    #list for subject with medians 84 x1
    list_sub_a.append(md_a)
    list_sub_b.append(md_b)
    list_sub_z.append(md_z)
    
    
# make arrays from list
alpha_md  = np.array(list_sub_a)
beta_md   = np.array(list_sub_b)
zeta_md   = np.array(list_sub_z)

np.save('C:/Users/canlab/Desktop/Modelling_Influenca/Results/Fit_all_subjects/Fitting_results/Weber_array_alpha_median_over_runs_all_sub.pkl',alpha_md)
np.save('C:/Users/canlab/Desktop/Modelling_Influenca/Results/Fit_all_subjects/Fitting_results/Weber_array_beta_median_over_runs_all_sub.pkl', beta_md)
np.save('C:/Users/canlab/Desktop/Modelling_Influenca/Results/Fit_all_subjects/Fitting_results/Weber_array_zeta_median_over_runs_all_sub.pkl', zeta_md)
```

```python
list_sub_a
```

Make list for parameter values in the form of 31 runs x n subjects for correlation

```python
list_a_runs=[]
list_b_runs=[]
list_z_runs=[]

list_=[]
total_runs=31
for run in range(total_runs):
    
    list_param_a = []
    list_param_b = []
    list_param_z = []
    
    # get param value from each sub in here
    
    list_.append(run)
    
    for sub in list_sub:
        
        str_sub = str(sub)
        max_run = np.max(data.loc[str_sub].index)
        
        # check for amount of runs
        if (run > max_run):
            #set value to nan
            list_param_a.append(np.nan)
            list_param_b.append(np.nan)
            list_param_z.append(np.nan)
            
        else:
        
            curr_sub = data.loc[str_sub, run]
            # appending the values for the parameters
            list_param_a.append(curr_sub['alpha'].tolist())
            list_param_b.append(curr_sub['beta'].tolist())
            list_param_z.append(curr_sub['zeta'].tolist())
     
    
    
    list_a_runs.append(list_param_a)
    list_b_runs.append(list_param_b)
    list_z_runs.append(list_param_z)
    
# make arrays from list
alpha=np.array(list_a_runs)
beta=np.array(list_b_runs)
zeta=np.array(list_z_runs)

np.save('C:/Users/canlab/Desktop/Modelling_Influenca/Results/Fit_all_subjects/Fitting_results/Weber_array_alpha_all_participants.pkl',alpha)
np.save('C:/Users/canlab/Desktop/Modelling_Influenca/Results/Fit_all_subjects/Fitting_results/Weber_array_beta_all_participants.pkl', beta)
np.save('C:/Users/canlab/Desktop/Modelling_Influenca/Results/Fit_all_subjects/Fitting_results/Weber_array_zeta_all_participants.pkl', zeta)

#wb_all=pd.concat(list_a_runs,keys=list_, names = ['Run', 'ID'])
#wh_all=pd.concat(list_b_runs,keys=list_, names = ['Run', 'ID'])
#nn_all=pd.concat(list_z_runs,keys=list_, names = ['Run', 'ID'])
```

Calulcate Spearman rank correlation

```python
total_runs=31

list_rho_a = []
list_p_a   = []

list_rho_b = []
list_p_b   = []

list_rho_z = []
list_p_z   = []


for run in range(total_runs):
    # spearman corr param values & medians over run
    # alpha
    sp_a = spearmanr(alpha[run], alpha_md, nan_policy='omit')
    list_rho_a.append(sp_a[0]) # append the rho values
    list_p_a.append(sp_a[1]) # append the p-values
    # beta
    sp_b = spearmanr(beta[run], beta_md, nan_policy='omit')
    list_rho_b.append(sp_b[0]) # append the rho values
    list_p_b.append(sp_b[1]) # append the p-values
    # zeta
    sp_z = spearmanr(zeta[run], zeta_md, nan_policy='omit')
    list_rho_z.append(sp_z[0]) # append the rho values
    list_p_z.append(sp_z[1]) # append the p-values
    
    
```

Make dataframes combining rho and p from correlation for parameters

```python
alpha_df = pd.DataFrame(list_rho_a)
alpha_df.columns=['rho']
alpha_df['p_value']=pd.DataFrame(list_p_a)
alpha_df.index.name = 'Run'
```

```python
beta_df = pd.DataFrame(list_rho_b)
beta_df.columns=['rho']
beta_df['p_value']=pd.DataFrame(list_p_b)
beta_df.index.name = 'Run'
beta_df
```

```python
zeta_df = pd.DataFrame(list_rho_z)
zeta_df.columns=['rho']
zeta_df['p_value']=pd.DataFrame(list_p_z)
zeta_df.index.name = 'Run'
zeta_df
```
